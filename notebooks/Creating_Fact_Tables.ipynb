```python
import pandas as pd
import numpy as np
```

```python
##imports
Categories = pd.read_csv('../data/Categories.csv')
Customers = pd.read_csv('../data/Customers.csv')
Employees = pd.read_csv('../data/Employees.csv')
OrderDetails = pd.read_csv('../data/OrderDetails.csv')
Orders = pd.read_csv('../data/Orders.csv')
Products = pd.read_csv('../data/Products.csv')
Shippers = pd.read_csv('../data/Shippers.csv')
Suppliers = pd.read_csv('../data/Suppliers.csv')
```

```python
print("Categories")
print(Categories.head())
print('------------------------------------------------------')
print("Customers")
print(Customers.head())
print('------------------------------------------------------')
print("Employees")
print(Employees.head())
print('------------------------------------------------------')
print("OrderDetails")
print(OrderDetails.head())
print('------------------------------------------------------')
print("Orders")
print(Orders.head())
print('------------------------------------------------------')
print("Products")
print(Products.head())
print('------------------------------------------------------')
print("Shippers")
print(Shippers.head())
print('------------------------------------------------------')
print("Suppliers")
print(Suppliers.head())
```

```python
# Fact Table: Orders
orders_fact_t = Orders.merge(Customers[['CustomerID', 'CustomerName', 'Country']], on='CustomerID') \
    .merge(Employees[['EmployeeID', 'FirstName', 'LastName']], on='EmployeeID') \
    .merge(Shippers[['ShipperID', 'ShipperName']], on='ShipperID')

orders_fact_t['EmployeeName'] = orders_fact_t['FirstName'] + ' ' + orders_fact_t['LastName']
orders_fact_t = orders_fact_t.drop(columns=['FirstName', 'LastName'])

# Fact Table: OrderDetails
orderdetails_fact_t = OrderDetails.merge(Orders[['OrderID']], on='OrderID') \
    .merge(Products[['ProductID', 'ProductName', 'Price', 'CategoryID']], on='ProductID')

# Fact Table: Products
products_fact_t = Products.merge(Categories[['CategoryID', 'CategoryName']], on='CategoryID') \
    .merge(Suppliers[['SupplierID', 'SupplierName']], on='SupplierID')
```
